```python
import numpy as np
import matplotlib.pyplot as plt

import datasets, utils

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def process(values): 
    return [float(x.strip()[:4]) for x in values.split('|')]

fb_normal = process("91.5 % | 96.0 % | 95.9 % | 96.3 % | 97.5 %")
fb_noisy = process("90.8 % | 96.1 % | 95.9 % | 65.8 % | 73.5 %")
fb_random = process("89.7 % | 57.1 % | 34.3 % | 38.2 % | 33.2 %")
ff_normal = process("91.5 % | 91.8 % | 89.8 % | 90.2 % | 93.3 %")
ff_noisy = process("91.1 % | 92.1 % | 90.4 % | 8.8 % | 7.0 %")
ff_random = process("90.0 % | 92.7 % | 91.0 % | 89.2 % | 91.5 %")

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height, '{:.0f} %'.format(height), ha='center', va='bottom')
        
width = 0.3
indices = np.arange(5)

plt.figure(figsize=(12, 2))
plt.axis('off')
rects1 = plt.bar(indices, ff_normal, width=width)
rects2 = plt.bar(indices+width, fb_normal, width=width)
autolabel(rects1)
autolabel(rects2)
plt.text(rects1[0].get_x() + rects1[0].get_width()/2., 5, 'FF', fontsize=18, color='white', ha='center', va='bottom')
plt.text(rects2[0].get_x() + rects2[0].get_width()/2., 5, 'FB', fontsize=18, color='white', ha='center', va='bottom')

plt.savefig('figures/without_noise.png', dpi=300)
```

```python
plt.figure(figsize=(12, 2))
plt.axis('off')
rects1 = plt.bar(indices, ff_noisy, width=width)
rects2 = plt.bar(indices+width, fb_noisy, width=width)
autolabel(rects1)
autolabel(rects2)
plt.text(rects1[0].get_x() + rects1[0].get_width()/2., 5, 'FF', fontsize=18, color='white', ha='center', va='bottom')
plt.text(rects2[0].get_x() + rects2[0].get_width()/2., 5, 'FB', fontsize=18, color='white', ha='center', va='bottom')

plt.savefig('figures/with_noise.png', dpi=300)
```
