```python
import numpy as np
import matplotlib.pyplot as plt

import datasets, utils

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def process(values): 
    return [float(x.strip()[:4]) for x in values.split('|')]

fb_normal = process("86.9 % | 92.9 % | 96.7 % | 96.8 % | 97.5 %")
fb_noisy = process("87.4 % | 92.8 % | 96.4 % | 66.9 % | 57.8 %")
ff_normal = process("86.3 % | 90.3 % | 93.6 % | 91.2 % | 87.9 %")
ff_noisy = process("85.6 % | 90.4 % | 93.3 % | 21.7 % | 21.7 %")

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height, '{:.0f} %'.format(height), ha='center', va='bottom')
        
width = 0.3
indices = np.arange(5)

plt.figure(figsize=(12, 3))
plt.axis('off')
rects1 = plt.bar(indices, ff_normal, width=width)
rects2 = plt.bar(indices+width, fb_normal, width=width)
autolabel(rects1)
autolabel(rects2)
plt.text(rects1[0].get_x() + rects1[0].get_width()/2., 5, 'FF', fontsize=18, color='white', ha='center', va='bottom')
plt.text(rects2[0].get_x() + rects2[0].get_width()/2., 5, 'FB', fontsize=18, color='white', ha='center', va='bottom')
```

```python
plt.figure(figsize=(12, 3))
plt.axis('off')
rects1 = plt.bar(indices, ff_noisy, width=width)
rects2 = plt.bar(indices+width, fb_noisy, width=width)
autolabel(rects1)
autolabel(rects2)
plt.text(rects1[0].get_x() + rects1[0].get_width()/2., 5, 'FF', fontsize=18, color='white', ha='center', va='bottom')
plt.text(rects2[0].get_x() + rects2[0].get_width()/2., 5, 'FB', fontsize=18, color='white', ha='center', va='bottom')
```
